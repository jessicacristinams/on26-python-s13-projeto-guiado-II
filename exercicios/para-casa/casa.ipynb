```python
# Tradução das colunas

traducoes = {
    'Location': 'Localização',
    'Number of Households (in 1000)': 'Número de famílias (1000)',
    'Percentage of households with pets': 'Porcentagem de famílias com Pets (%)',
    'Number of Pet Households (in 1000)': 'Número de famílias com Pets (1000)',
    'Percentage of Dog Owners': 'Porcentagem de donos de cães (%)',
    'Dog Owning Households (1000s)': 'Famílias com cães (milhares)',
    'Mean Number of Dogs per household': 'Média de cães por famílias',
    'Dog Population (in 1000)':  'População canina (1000)',
    'Percentage of Cat Owners': 'Porcentagem de donos de gatos (%)',
    'Cat Owning Households': 'Famílias com gatos',
    'Mean Number of Cats' : 'Média de gatos por famílias',
    'Cat Population': 'População felina'
}

# result = [(d['color'], d['value']) for d in dictionarylist]

novas_colunas = []
for chave, valor in traducoes.items():
    novas_colunas.append(valor)

print(novas_colunas)
```

```python
# atribuindo os novos nomes das colunas
df.columns = novas_colunas
df.head(10)
```

```python
#para ver as informações do dataframe, quantidade de linhas e colunas
df.shape
```

```python

# Obter o valor máximo e mínimo do número de famílias com Pets:
valor_max = df['Número de famílias com Pets (1000)'].max()
valor_min = df['Número de famílias com Pets (1000)'].min()

print("Valor máximo do número de famílias com Pets (1000)")
print(valor_max)

print("Valor mínimo do número de famílias com Pets (1000)")
print(valor_min)


```

```python


df[['Número de famílias com Pets (1000)']].describe(percentiles=[0.99]).T

```

**TOP 5 CIDADES COM MAIS PETS**

```python
# 5 cidades que contém o maior número de famílias com pets

top5_cidades = df.loc[:, ["Localização", "Número de famílias com Pets (1000)"]].sort_values(by='Número de famílias com Pets (1000)', ascending=False).head(5)
top5_cidades

```

```python
top5_cidades.plot(kind='bar')


plt.xlabel('Cidades')
plt.ylabel('N de Famílias com pets')
plt.title('Top 5 Cidades / Número de famílias com pets')

plt.show()
```

**5 CIDADES AS FAMÍLIAS TÊM MAIS CÃES**

```python
# 5 cidades com mais famílias com gatos
top5_gatos = df.loc[:, ["Localização", "Famílias com gatos"]].sort_values(by='Famílias com gatos', ascending=False).head(5)
top5_gatos
```

```python
# 5 cidados com mais famílias com cães
top5_caes = df.loc[:, ["Localização", "Famílias com cães (milhares)"]].sort_values(by='Famílias com cães (milhares)', ascending=True).head(5)
top5_caes
```
